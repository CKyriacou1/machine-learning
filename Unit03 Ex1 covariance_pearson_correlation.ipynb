```python
# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
#Seed random number generator
seed(1)
```

```python
#Prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)
```

```python
data1
```

```python
data2
```

```python
#Calculate covariance matrix
covariance = cov(data1, data2)
covariance
#The covariance between the two arrays is 389.7545618 which indicates a positive(duw to the + sign) and strong(due to the size of the number) correlation
```

```python
#Calculate Pearson's correlation
corr= pearsonr(data1, data2)
corr
#The Pearson's correlation coefficient between the two arrays is 0.8876119085795297 which indicates positive and strong correlation as well
#The 0.0 value indicates the p-value known as level of significance.This is the chance of finding a correlation coefficient of 0.887 or even higher in a sample if in the population it would be 0(no association) 
```

```python
#Plot
plt.scatter(data1, data2)
plt.show()
#As we can see from the scatter plot the is a strong postitive linear relationship between data 1 and data 2.
#That is the variables are moving in the same direction(while values in data 1 increases values in data 2 increases as well and the opposite)
```

```python
#Summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr[0])
```
